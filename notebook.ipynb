```python
import ast

def convert(obj):
    if isinstance(obj, str):              
        try:
            obj = ast.literal_eval(obj)   
        except (ValueError, SyntaxError):
            return []  
    L = []
    
    for i in obj[:3]:  
        if isinstance(i, dict) and 'name' in i:  
            L.append(i['name'])
    return L

movie['cast'] = movie['cast'].apply(convert)


```

```python
import ast

def convert_crew(obj):
    if isinstance(obj, str):       # if it's a string
        obj = ast.literal_eval(obj)  # convert to list of dicts
    
    for member in obj:
        if member['job'] == 'Director':
            return [member['name']]  # return as a list
    
    return []  # empty list if no director

# Apply to crew column
movie['crew'] = movie['crew'].apply(convert_crew)
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []
    for word in text.split():      # Split text into words
        y.append(ps.stem(word))    # Stem each word
    return " ".join(y)             # Join back into a string
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words ='english')
```

```python
def recommend(movie):
    movie_index = new_dataframe[new_dataframe['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse =True,key = lambda x: x[1])[1:6]

    for i in movie_list:
        print(new_dataframe.iloc[i[0]].title)
```
